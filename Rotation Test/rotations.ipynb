```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# We are using the mx (x-axis rotation) dimension from ORN
df = pd.read_csv('Sensor testing/Rotation testing_ORN.csv')
df
```

```python
# Plot it over time to see that we rotate it over time!
x = df["TimeStamp"]
# Start time at x=0
x = x - x[0]
y = df["mx"]
plt.plot(x, y)
plt.grid()
plt.xlabel("Time")
plt.ylabel("MX Rotation units")
plt.title("Rotation over time")
plt.show()
```

Looking at the above graph, there are 3 specific rotation points. I would have thought there are 4, as I rotated it 90 degrees each time, but this should be fine. I will just detect if it is in the lowest step or not.

```python
threshold = y > 300

plt.plot(x, threshold, label="Threshold")
plt.grid()
plt.xlabel("Time")
plt.ylabel("MX Rotation units")
plt.title("Rotation over time")
plt.show()
```
