# Understanding the GAPED procedure
In the GAPED dataset we're provided with ~700 pictures with *Arousal* and *Valence* scores attached to them. We plotted each dataset on the scatter plot the results of which you can see below, as well as our initial expectations




<img src="https://www.mdpi.com/sensors/sensors-23-01598/article_deploy/html/images/sensors-23-01598-g001.png" style="width: 300px;"/>

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

def generateplot(filename):
    # reading the csv file
    df = pd.read_csv(filename, delimiter='\t')
    plt.scatter(df['Valence'], df['Arousal'])
    
    #getting the line of best fit
    m,b = np.polyfit(df['Valence'], df['Arousal'], 1)
    x_axis = np.linspace(min(df['Valence'])-20, max(df['Valence'])+20, num =100)
    plt.plot(x_axis,m*x_axis + b, color = 'r')
    
    #plotting the points for the file
    plt.title(f'Ratings of images {filename[-5:-4]}')
    
    #setting the axis right
    plt.xlim((0,100))
    plt.ylim((0,100))
    plt.xlabel('Valence')
    plt.ylabel('Arousal')
    plt.show()

#applying the function to data sets
link = 'https://raw.githubusercontent.com/MyelinGroup/myelin-fuzzy-logic/main/GAPED%20Procedure/'
generateplot(link + 'P.txt') #Positive emotion
generateplot(link + 'A.txt') #Agressive emotion
generateplot(link + 'H.txt') #Happy emotion
generateplot(link + 'N.txt') #Neutral emotion
```

# Conclusion
Emotions are not induced correctly using the pictures, as they're grouping differently from the prediction. Moreover, most of them lie on anger-boredom line.
