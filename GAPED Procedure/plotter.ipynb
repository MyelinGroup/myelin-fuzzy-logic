```python
import matplotlib.pyplot as plt
import pandas as pd

def generateplot(filename):
    df = pd.read_csv(filename, delimiter='\t')

    plt.scatter(df['Valence'], df['Arousal'])

    for i in range(len(df)):
        name = df['Name'][i]
        name = name.split('.')[0]
        plt.annotate(name, (df['Valence'][i], df['Arousal'][i]))

    plt.title(f'Ratings of images for {filename[0]}')
    plt.xlabel('Valence')
    plt.ylabel('Arousal')
    plt.show()

generateplot('P.txt')
generateplot('A.txt')
generateplot('H.txt')
generateplot('N.txt')
```
